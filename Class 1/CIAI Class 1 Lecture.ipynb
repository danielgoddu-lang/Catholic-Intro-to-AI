## 2. Define Artificial Intelligence (AI) in your own words.

What is AI to you? Define AI in your own words.


### 2A. What is Artificial Intelligence (AI)?

* Definition: AI is the field of computer science that focuses on creating systems capable of performing tasks that typically require human intelligence. This includes problem-solving, understanding language, recognizing patterns, and making decisions.

* Examples: Siri, Google Translate, facial recognition, Netflix recommendations, email spam filters
* Misconception: AI is not conscious, sentient, or creative. Most current AI is "narrow" or task-specific.


### 2B: AI Capabilities

* `Narrow` AI that is good at one specific task (or a narrow set of tasks)

    * Examples are: spam filters, fraud detection models, recommendation systems, voice assistants, image classifiers, generative AI (ChatGPT, Copilot, Claude, etc.)
    * Each of these systems can only perform well in its own domain
    * Most AI systems today are `narrow` or `weak` AI
* `General` AI can learn and performan any intellectual taks a human can
    * Not a pre-programmed task
    * Examples are: learn a new job without training, reason across domains, transfer knowledge
* `Superintelligence` AI that is more capable that humans at essentially all intellectual tasks
    * Can reason, create, discover, strategize, and socially understand all on its own
    * Far beyond humans


## 3. Distinguish between Computer Science, AI, Machine Learning (ML), and Deep Learning (DL).

AI is the broad concept of machines doing intelligent things.

- Machine Learning (ML) is a subset of AI that uses data and algorithms to "learn" from experience using `labels`.

- Deep Learning (DL) is a further subset of ML that uses artificial neural networks to model complex patterns.
- Analogy: Think of AI as the universe, ML as a galaxy, and DL as a solar system within that galaxy.
- Supervised Learning is a type of machine learning where a model is trained using **labeled data** — that is, data where the correct answers are already known. In this approach, the model learns by comparing its predictions to the correct answers and gradually improving over time. For example, if you train a model with many emails labeled “spam” or “not spam,” it learns patterns that help it classify new emails correctly. *You show the computer examples with the right answers, and it learns to predict those answers on its own.*

- Unsupervised Learning is a type of machine learning where a model is given **unlabeled data** and must find patterns, groupings, or structure on its own. Unlike supervised learning, no one tells the model the “right answer” ahead of time. Instead, it looks for similarities or hidden relationships in the data. For example, a model might group customers into different categories based on buying behavior without being told what those categories should be. *The computer is given a pile of data and asked to discover what patterns are there by itself.*

This diagram shows how several important technology fields are related to one another. The biggest circle is **Computer Science**, which is the broad field that includes programming, software, hardware, networks, databases, cybersecurity, and much more. Inside computer science is **Artificial Intelligence**, which focuses specifically on building systems that can do things that seem intelligent, such as recognizing speech, making decisions, or understanding language.

Inside artificial intelligence is **Machine Learning**, which is a more specific area where computers learn patterns from data instead of being told every rule directly by a programmer. For example, instead of programming every rule for identifying spam email, a machine learning system studies many examples of spam and non-spam messages and learns the patterns for itself.

Inside machine learning is **Deep Learning**, which is an even more specialized approach that uses layers of artificial “neurons” called neural networks. Deep learning is what powers many of today’s most advanced AI systems, such as image recognition, voice assistants, and large language models like ChatGPT.

The **Data Science** circle overlaps with AI and machine learning because these fields often use the same tools and data. However, data science is not exactly the same as AI. Data science is more focused on collecting, organizing, analyzing, and drawing conclusions from data. A data scientist might use machine learning, but they might also just use statistics, charts, and spreadsheets to answer questions.

The main idea students should understand is this: **AI is part of computer science, machine learning is part of AI, and deep learning is part of machine learning.** Data science overlaps with all of them because data is essential to how modern AI works.
