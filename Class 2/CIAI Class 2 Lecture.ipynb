# Welcome to A Catholic Introduction to Artificial Intelligence Class 2!

“The question is not whether intelligent machines can have any emotions, but whether machines can be intelligent without any emotions.” – Marvin Minsky

## Learning Objectives 

By the end of this class, students will be able to:

1.	Describe the historical roots of AI and automation.

2.	Identify key figures and milestones in early AI development.
3.	Understand the difference between traditional automation and intelligent systems.
4.	Explain the concept of the "AI Winter" and its significance.


## 1. Class 1 Review




- "What is AI?"

- "What do you think the first AI looked like?"


## 2. Early Automation and the Roots of AI


- Introduce early examples: Jacquard Loom, assembly lines.
- Explain how machines were used to replicate repetitive tasks long before computers.
- Introduce Alan Turing and the Turing Test.
- Optional: Show a brief clip on Turing (e.g., CrashCourse AI or History.com).


## 3. The First Generation of AI and its Challenges.

- Identify key figures and milestones in early AI development.
- Describe ELIZA and early expert systems.
- Explain symbolic AI (rule-based logic).
- Group Activity: Read and discuss a transcript from ELIZA.


## 5. Explain the concept of the "AI Winter" and its significance.

- Define "AI Winter" and why it happened.
- Discuss overpromising and underdelivering.
- Show a simple chart or graph of AI hype over time.
- Tie into today: What’s different now? (More data, better hardware)
